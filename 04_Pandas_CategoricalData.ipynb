Categorical Data:


*   Categorical data in Pandas refers to data that takes on a limited number of fixed values, often strings or labels.

Uses:


*   Memory-intensive



*  Slow to filter, group, or compare

   
Syntax:



*   .astype('category')











```python
import pandas as pd

df = pd.DataFrame({
    'Country': ['USA', 'India', 'USA', 'Germany', 'India', 'USA']
})

# Before conversion
print("Before:")
print(df.dtypes)
print(df['Country'].memory_usage(deep=True), "bytes")

# Convert to category
df['Country'] = df['Country'].astype('category')

print("\nAfter:")
print(df.dtypes)
print(df['Country'].memory_usage(deep=True), "bytes")
```

```python
#groupby() performance
result = df.groupby('Country').size()
print(result)
```
