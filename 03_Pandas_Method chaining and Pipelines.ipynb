Method Chaining and Pipelines

* Method chaining in Pandas means calling multiple DataFrame methods in a single line, like a chain  




*   .pipe() is a special method that lets you inject a function into a method chain.






.assign() - add/modify cols

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({

    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [80, 70, 90],
    'Science': [85, 60, 95]
})

# Add a new column 'Average'
df = df.assign(Average=lambda d: (d['Math'] + d['Science']) / 2)
print(df)
```

.querry() - SQL like Filtering

```python
df.query('Average > 85')
```

.pipe() - Apply Custom Changes


```python
def capitalize_names(df):                        #Capiitalize_names defines a custom Python function named
    df['Name'] = df['Name'].str.capitalize()
    return df

df = df.pipe(capitalize_names)
print(df)
```

```python
# example of full .pipe()
df = (
    pd.DataFrame({
    'Name': ['alice', 'BOB', 'ChaRLIE'],
    'Math': [80, 70, 90],
    'Science': [85, 60, 95]
})
    .assign(Average=lambda d: (d['Math'] + d['Science']) / 2)
    .query('Average > 85')
    .pipe(capitalize_names)
)

df
```
